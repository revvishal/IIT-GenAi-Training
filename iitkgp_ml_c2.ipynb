```python
sepal_length = np.array(dataset['sepal_length'])
sepal_width = np.array(dataset['sepal_width'])
petal_length = np.array(dataset['petal_length'])
petal_width = np.array(dataset['petal_width'])
```

```python
mean_sepal_length = np.mean(sepal_length)
std_sepal_width = np.std(sepal_width)
min_petal_length = np.min(petal_length)
max_petal_width = np.max(petal_width)

print("mean sl:",mean_sepal_length)
print("std sw:",std_sepal_width)
print("min pl:",min_petal_length)
print("max pw:",max_petal_width)
```

```python
large_sepals = sepal_length > mean_sepal_length
print(large_sepals)
print(f'Flowers with large sepals: {sepal_length[large_sepals]}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data = dataset, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
sns.pairplot(dataset, hue='species')
plt.show()
```

```python
corr_mat = dataset.corr(numeric_only=True)
print(corr_mat)
```

```python
sns.heatmap(corr_mat, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```
